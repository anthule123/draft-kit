```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Khởi tạo biểu đồ
fig, ax = plt.subplots(figsize=(14,8))
ax.set_aspect('equal')
ax.axis('off')
ax.set_xlim(-1.5, 7)
ax.set_ylim(-2, 2)
# Vẽ Trái Đất (hình tròn)
R=1.5
earth = patches.Circle((0, 0), radius=R, edgecolor='blue', facecolor='lightblue', linewidth=2)
ax.add_patch(earth)
ax.text(0,  - R - 0.3, 'Trái Đất', ha='center', fontsize=10, color='blue')

# # Tọa độ 2 thành phố: Syene (dưới) và Alexandria (trên, lệch góc 7.2 độ)
city1 = (R, 0)  # Syene (ngay dưới tâm)
angle_deg = 7.2
angle_rad = np.deg2rad(angle_deg)
city2 = (R * np.cos(angle_rad), R * np.sin(angle_rad))  # Alexandria

# # Vẽ hai thành phố
ax.plot(*city1, 'ko')  # Syene
ax.text(city1[0] - 0.1, city1[1] - 0.2, 'Syene', fontsize=10)

ax.plot(*city2, 'ko')  # Alexandria
ax.text(city2[0] + 0.1, city2[1] + 0.1, 'Alexandria', fontsize=10)

# Vẽ đường nối hai thành phố về tâm Trái Đất
ax.plot([0, city2[0]], [0, city2[1]], 'k--', linewidth=1)
ax.plot([0, city1[0]], [0, city1[1]], 'k--', linewidth=1)

# Vẽ góc α = 7.2°
arc = patches.Arc((0, 0), 1, 1, theta1=0, theta2=0 + angle_deg, color='red')
ax.add_patch(arc)
ax.text(0.5, 0.15, r"$\alpha = 7.2^\circ$", color='red', fontsize=10)


# Vẽ Mặt Trời (hình tròn bên phải)
sun_center = (5, 0)
sun_radius = R
sun = patches.Circle(sun_center, radius=sun_radius, edgecolor='orange', facecolor='gold', linewidth=2)
ax.add_patch(sun)
ax.text(sun_center[0], sun_center[1] - sun_radius - 0.3, 'Mặt Trời', ha='center', fontsize=10, color='orange')

# Vẽ các tia sáng song song từ Mặt Trời đến Trái Đất (5 tia)
# Tia giữa đi qua Syene (gần tâm Trái Đất)
y_offsets =  np.linspace(-2*angle_rad, 2* angle_rad, 5)
for y in y_offsets:
    ax.arrow(sun_center[0] - sun_radius * np.cos(y), y, 
              -2.1  +sun_radius * np.cos(y)  , 0,
             head_width=0.05, head_length=0.08,  fc='orange', ec='orange', linestyle="--")
 

plt.title("Minh họa thí nghiệm Eratosthenes đo chu vi Trái Đất", fontsize=12)
plt.savefig('geometry.svg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Khởi tạo biểu đồ
fig, ax = plt.subplots(figsize=(14,8))
ax.set_aspect('equal')
ax.axis('off')
ax.set_xlim(-1.5, 3)
ax.set_ylim(-2, 2)
# Vẽ Trái Đất (hình tròn)
R=1.5
earth = patches.Circle((0, 0), radius=R, edgecolor='blue', facecolor='lightblue', linewidth=2)
ax.add_patch(earth)
ax.text(0,  - R - 0.3, 'Trái Đất', ha='center', fontsize=10, color='blue')

# # Tọa độ 2 thành phố: Syene (dưới) và Alexandria (trên, lệch góc 7.2 độ)
city1 = (R, 0)  # Syene (ngay dưới tâm)
angle_deg = 7.2
angle_rad = np.deg2rad(angle_deg)
city2 = (R * np.cos(angle_rad), R * np.sin(angle_rad))  # Alexandria

# # Vẽ hai thành phố
ax.plot(*city1, 'ko')  # Syene
ax.text(city1[0] - 0.1, city1[1] - 0.2, 'Syene', fontsize=10)

ax.plot(*city2, 'ko')  # Alexandria
# ax.text(city2[0] + 0.1, city2[1] + 0.3, 'Alexandria', fontsize=10)

# Vẽ đường nối hai thành phố về tâm Trái Đất
ax.plot([0, city2[0]], [0, city2[1]], 'k--', linewidth=1)
ax.plot([0, city1[0]], [0, city1[1]], 'k--', linewidth=1)

# Vẽ góc α = 7.2°
arc = patches.Arc((0, 0), 1, 1, theta1=0, theta2=0 + angle_deg, color='red')
ax.add_patch(arc)
ax.text(0.5, 0.15, r"$\alpha = 7.2^\circ$", color='red', fontsize=10)


# Mô tả hình học - góc so le trong
x_points = [2*R*np.cos(angle_rad), R]
y_points = [0, 0]
ax.plot(x_points,y_points, color='orange') 

x_points = [R*np.cos(angle_rad), 2*R*np.cos(angle_rad)]
y_points = [ R* np.sin(angle_rad), 2* R* np.sin(angle_rad)]
ax.plot(x_points,y_points, color='black', linestyle="--")

x_points = [2* R*np.cos(angle_rad), R*np.cos(angle_rad)]
y_points = [2* R* np.sin(angle_rad), 2* R* np.sin(angle_rad)]
ax.plot(x_points,y_points, color='orange')

# Vẽ góc α = 7.2° so le
arc = patches.Arc((2* R*np.cos(angle_rad), 2* R* np.sin(angle_rad)), 1, 1, theta1=180, theta2=180+angle_deg, color='red')
ax.add_patch(arc)


plt.title("Góc giữa 2 đường nối 2 thành phố đến tâm Trái Đất = góc giữa tia nắng và cột", fontsize=12)
plt.savefig('geometry2.svg')
plt.show()


```
